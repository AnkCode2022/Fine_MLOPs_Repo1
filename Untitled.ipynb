```python
import time

data_monitor = DefaultModelMonitor.attach("Drift-Monitor")

while True:
    executions = data_monitor.list_executions()
    if executions:
        latest = executions[-1].describe()
        status = latest["ProcessingJobStatus"]
        print(f"📋 Status: {status}")

        if status in ["Completed", "CompletedWithViolations", "Failed", "Stopped"]:
            print(f"✅ Job finished with: {status}")
            print(f"💬 Message: {latest.get('ExitMessage', 'N/A')}")
            break
    else:
        print("⏳ No executions yet...")

    time.sleep(30)
```

```python
import pandas as pd
import io
import json

# Read captured file
obj = s3.get_object(
    Bucket="placement-project-bkt",
    Key="monitoring/data-capture/Driftplacement-model-endpointt/AllTraffic/2026/04/05/17/45-38-843-253036a3-cffc-4f07-86ea-514ad8483499.jsonl"
)
content = obj["Body"].read().decode("utf-8")

# Simulate preprocessor on each line
print("📊 What preprocessor will extract:\n")
rows = []
for line in content.strip().split("\n"):
    record = json.loads(line)
    csv_data = record["captureData"]["endpointInput"]["data"].strip()
    df = pd.read_csv(io.StringIO(csv_data), header=None)
    df = df.iloc[:, :3]
    df.columns = ["cgpa", "iq", "profile_score"]
    rows.append(df)
    print(df.to_string(index=False))
    print("-" * 40)

# Final combined dataframe
final_df = pd.concat(rows, ignore_index=True)
print(f"\n✅ Total rows    : {len(final_df)}")
print(f"✅ Total columns : {len(final_df.columns)}")
print(f"✅ Columns       : {list(final_df.columns)}")
print(f"\n📊 Stats:\n{final_df.describe()}")
```
